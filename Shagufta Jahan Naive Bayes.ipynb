# Gaussian Naive Bayes

```python
# training the model on training set
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(train_X,train_Y)
```

```python
y_pred = gnb.predict(train_X)
y_pred
```

```python
from sklearn import metrics
print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(train_Y, y_pred))
print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(train_Y, y_pred)*100)
```

# Gaussian Naive Bayes accuracy of train data set is 79.53%

```python
# making predictions on the testing set
y_pred = gnb.predict(test_x)
```

```python
y_pred
```

```python
# comparing actual response values (y_test) with predicted response values (y_pred)
from sklearn import metrics
print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(test_y, y_pred)*100)
```
